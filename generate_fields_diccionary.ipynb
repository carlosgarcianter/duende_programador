```python
def formatter(fields):

  processed_matched_fields = []
  for field in fields:
      lowercase_field = field.lower()
      processed_field = re.sub(r'[^a-z0-9_]+', '_', lowercase_field)
      processed_matched_fields.append(processed_field)
  return processed_matched_fields
```

```python
def format_field_to_title_case(field, acronyms=set()):

  words = field.split('_')

  processed_words = []
  for word in words:
    if word.lower() in acronyms:
      processed_words.append(word.upper())
    else:
      processed_words.append(word.capitalize())

  formatted_field = '_'.join(processed_words)

  return formatted_field

special_acronyms = {'dc2', 'dc3', 'fx', 'dc4', 'dc7', 'de1', 'de2', 'de3', 'pl',
                    'eur', 'bo', 'dc', 'de', 'dc1', 'boco', 'isin', 'dc13',
                    'dop', 'sla', 'cpty', 'ge', 'rof', 'fobo',
}
```
